Mini Project One: An Analysis on Fast Food Consumption

This Data will cover how often do people eat fast food, and what factors influence when, where, and why they choose it? First, Let's import our libraries 

```python
import pandas as pd
import matplotlib.pyplot as plt
import csv
```

Next, I'll import my csv where I captured my responses(12)

```python
df = pd.read_csv('Analysis of Fast Food Consumption.csv')

#First few rows of my data
df.head()

#Checking the shape of my data
df.shape

# Checking for any missing data
df.isnull().sum()
```

This will be my WHO Section (Demographics). I am going to begin with calculatng percentages for Gender

```python
gender_counts = df['Gender'].value_counts()
gender_percentages = (df['Gender'].value_counts(normalize=True) * 100).round(2)

gender_counts
```

```python
gender_percentages
```

```python
# Here is a Visualization for Gender to have a better understanding of my data
#  (Also if you don't feel like reading, I will include visuals for the rest of my data as well)

df['Gender'].value_counts().plot(kind='bar')
plt.title('Gender')
plt.xlabel('Gender')
plt.ylabel('Number of People')
plt.xticks(rotation=0)
plt.show()
```

Next, I am going to calcuate percetages for Year in School

```python
year_counts = df['Year in School'].value_counts()
year_percentages = (df['Year in School'].value_counts(normalize=True) * 100).round(2)

year_counts
```

```python
year_percentages
```

```python
#Visualization for Year in School
df['Year in School'].value_counts().plot(kind='bar')
plt.title('Year in School')
plt.xlabel('Year')
plt.ylabel('Number of People')
plt.xticks(rotation=45)
plt.show()
```

Here is my WHERE section. I am Calculating where do you get fast food.
